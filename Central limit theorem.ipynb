```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
die=pd.Series([1,2,3,4,5,6])
# roll 5 times
sample_5 = die.sample(5, replace=True)
print(np.std(sample_5))
```

```python
# get 10 means from each roll five times
lista = []
for i in range(10):
    sample_5 = die.sample(5, replace=True)
    
    lista.append(np.std(sample_5)) 
```

```python
print(lista, 'Mean: ',np.mean(lista))
```

```python
plt.hist(lista,bins=10)
plt.axvline(x=np.mean(lista), color='r', linestyle='dashed', linewidth=2)
plt.show()
```
