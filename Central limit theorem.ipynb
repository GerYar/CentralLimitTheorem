```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Defining a function to plot a histogram

def plotting_hist(lista,c):    
    fig, ax = plt.subplots()
    fig.set_size_inches([7,2])
    ax.hist(lista, color=c, edgecolor='black')
    plt.axvline(x=np.mean(lista), color='r', linestyle='dashed', linewidth=2, label='Mean= '+str(round(np.mean(lista),3)))
    ax.set_xlabel('Mean of throw 5 times a die')
    ax.set_ylabel('Frequency')
    ax.set_title(str(sample)+' repetitions')
    ax.legend()
    plt.show()
```

## Discret uniform distribution

In a fair die the probability of get each value is equal 1/6.

```python
#plotting

valores_x = np.arange(1, 7)
probabilidades = np.array([1/6, 1/6, 1/6, 1/6, 1/6, 1/6])
plt.bar(valores_x, probabilidades, color='blue')
plt.xlabel("Die value")
plt.ylabel("Probability")
plt.title("Probability of each value in a die")
plt.show()
```

Calculating a sample of a discrete distribution

```python
# All posible results throwing a die 
die=pd.Series([1,2,3,4,5,6])

# 5 results throwing a die 5 times
sample_5 = die.sample(5, replace=True)
sample_5
```

We will throw the die 5 times and each time we will get the mean, we will repeat this process:
* 10 times
* 100 times
* 1000 times

expecting that the sampling distribution of a statistic becomes closer to the normal distribution as the number of trials increases

```python

# number of repetitions
samples=[10,100,1000]
for sample in samples:
    
# each try
    means_list = []
    for i in range(sample):
        sample_5 = die.sample(5, replace=True)
        
        #getting the mean of all tries
        means_list.append(np.mean(sample_5))

    # Calling funtion to plot
    plotting_hist(means_list, 'blue')
```

As we can see if we have more tries, the distribution looks as a normal distribution

## Poisson distribution

Let's suppose we want to simulate the number of phone calls that a customer service center receives in an hour. We know that the average number of calls per hour is 3.

Generate 1000 random numbers from the Poisson distribution with mean 3.

```python
# importing Poisson distribution
from scipy.stats import poisson

# generatind 1000 samples
sample_poisson = poisson.rvs(3, size=1000)

# ploting samples
plt.hist(sample_poisson, edgecolor='black', color='yellow')
plt.show()
```

```python
# number of repetitions
samples=[10,100,1000]

for sample in samples:
    
# get 10000 means from each five times
    means_samples = []
    for i in range(sample):
        sample_1000 = poisson.rvs(3, size=1000)
        
        #getting the mean of all tries
        means_samples.append(np.mean(sample_1000))
    
    # Calling funtion to plot
    plotting_hist(means_samples, 'yellow')   
    
```

As we can see again if we have more tries, the distribution looks as a normal distribution

### The Central Limit Theorem is a powerful concept that allows us to understand how the means of random samples behave as the sample size increases, regardless of the shape of the original distribution. This is fundamental in statistical inference and in the construction of confidence intervals and hypothesis testing.
